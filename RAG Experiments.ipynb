```python
import trafilatura
import requests
from bs4 import BeautifulSoup
import re
```

```python
from pydoc import html


def extract_text_trafilatura(html_content):
    extracted_text = trafilatura.extract(html_content)
    return extracted_text



def extract_using_bs_main(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    extracted_text = soup.find('main')
    if extracted_text:
        extracted_text = extracted_text.get_text(separator='\n')
        return extracted_text
    else:
        return 


def clean_text(s: str):
    s = re.sub(r"\s+", " ", s).strip()
    return s

def split_paragraphs(text: str):
    paras = [clean_text(p) for p in re.split(r"\n", text) if clean_text(p)]
    return paras
```

```python
def process_url(url):

    if((url is None) | (url.strip() == "" )):
        print("URL is Invalid. Please provide a valid URL.")
        return
    
    response = requests.get(url)
    if response.status_code == 200:
        html_content = response.text
    else:
        print(f"Failed to retrieve the webpage. Status code: {response.status_code}")


    extracted_text = extract_using_bs_main(html_content)
    if(extracted_text is None):
        extracted_text = extract_text_trafilatura(html_content)

    paras = split_paragraphs(extracted_text)

    return paras
```

```python
process_url("https://stripe.com/in/guides/payment-methods-guide")
```

```python
from aiohttp import ClientSession, ClientTimeout,TCPConnector
import asyncio

CONCURRENCY = 12
TIMEOUT = 25
url_link = "https://stripe.com/in/guides/payment-methods-guide"

DEFAULT_TIMEOUT =  ClientTimeout(
                total=None,
                sock_connect=TIMEOUT,
                sock_read=TIMEOUT
                
)
## TCP Connection
conn = TCPConnector(limit=CONCURRENCY)
## Define Timeout
# timeout = DEFAULT_TIMEOUT

async with ClientSession(
connector=conn,
timeout=DEFAULT_TIMEOUT,

)as session:
    sem = asyncio.Semaphore(CONCURRENCY)

    async with sem:
        print(process_url(url_link))
```
